```python
import copy
original =  [[1,2],[2,4]]

shallow = copy.copy(original)
#deep = copy.deepcopy(original)
original[0][0] = 99

print(original)  # [[99, 2], [3, 4]]
print(shallow)   # [[99, 2], [3, 4]]  ← changed too!
#print(deep)

print(id(original))
print(id(shallow))
#print(id(deep))


print(id(original[1]))
print(id(shallow[1]))
```

**Python interns small integers (usually from -5 to 256)**

```python
a = "hello"
b = "hello"

print(a is b)   # True

a = 10
b = 10

print(a is b)   # True

a = "hello world"
b = "hello world"

print(a is b)   # Sometimes False

#why ? Strings with spaces or created dynamically may NOT be interned automatically
```

**Forcing Interning**

```python
import sys

a = sys.intern("hello world")
b = sys.intern("hello world")

print(a is b)   # True
```

```python
a = "hi"
b = "hi"

print(a == b)  # True (values same)
print(a is b)  # True (same object due to interning)

# == → checks value
# is → checks memory (identity)
```
