```python
# Bitwise NOT Op -
# The ~ operator performs a bitwise NOT (inversion) on an integer — it flips all the bits:
# 1 → 0
# 0 → 1

x = 5        # binary: 00000101
result = ~x  # binary: 11111010

print(result)  # Output: -6

# Python uses two’s complement representation for integers. The rule is:

# ~0   # -1
# ~1   # -2
# ~10  # -11
```

```python
# a << b shifts the binary representation of a to the left by b positions.
# Each left shift is equivalent to multiplying by 2.

print( 3 << 1 )

#Step-by-step

# 1) Convert 3 to binary:
#    3 = 00000011

# 2) Shift left by 1:
#    00000011 << 1 → 00000110

# 3) Convert back to decimal:
#    00000110 = 6


# a >> b shifts the binary representation of a to the right by b positions.
# Each right shift is equivalent to multiplying by 2.

print( 3 >> 1 )

```

#Extract values from list with two methods
1) indexing

2) slicing

    a. start = include

    b. end = n - 1

    c. step = sequence


```python
#--------------->      0       1        2      3
names : list[str] = ["Qasim","Rehan","Ameen","Junaid"]
#<--------------        -4      -3     -2      -1
# slicing
names[0:2]
```

```python
#--------------->      0       1        2      3
names : list[str] = ["Qasim","Rehan","Ameen","Junaid"]
#<--------------        -4      -3     -2      -1
# slicing - always left to right
print(names[-3:-1])

print(names[1:-1])
```
