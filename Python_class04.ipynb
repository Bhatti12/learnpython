**Replacing & Translating**

```python
s = "hello"

print(s.replace("l", "x"))   # "hexxo"

table = str.maketrans("h", "j")
print(table)
print(s.translate(table))    # "jello"
```

**Splitting & Joining**

```python
s = "a,b,c"

print(s.split(","))        # ['a', 'b', 'c']
print(s.rsplit(",", 1))   # ['a,b', 'c']

print("a\nb\nc".splitlines())  # ['a', 'b', 'c']

print("-".join(["a", "b", "c"]))  # "a-b-c"
```

**Stripping & Removing**

```python
s = "  hello  "

print(s.strip())        # "hello"
print(s.lstrip())       # "hello  "
print(s.rstrip())       # "  hello"

print("unhappy".removeprefix("un"))  # "happy"
print("file.txt".removesuffix(".txt")) # "file"
```

**Checking Content**

```python
print("abc123".isalnum())     # True
print("abc".isalpha())        # True
print("123".isdecimal())      # True
print("123".isdigit())        # True
print("123".isnumeric())      # True
print("abc".islower())        # True
print("ABC".isupper())        # True
print("Hello World".istitle())# True
print("   ".isspace())        # True
print("abc".isascii())        # True
print("var_1".isidentifier()) # True
print("abc!".isprintable())   # True
```

**Searching & Counting**

```python
s = "banana"

print(s.count("a"))        # 3
print(s.find("na"))        # 2
print(s.rfind("na"))       # 4
print(s.index("na"))       # 2 (error if not found)
print(s.rindex("na"))      # 4
```

**Alignment & Padding**

```python
s = "hi"

print(s.center(10, '-'))   # "---hi-----"
print(s.ljust(10, '-'))    # "hi--------"
print(s.rjust(10, '-'))    # "--------hi"
print("42".zfill(5))       # "00042"
```

**Basic Transformations**

```python
s = "hello world"

print(s.capitalize())   # "Hello world"
print(s.casefold())     # "hello world" (stronger lowercase)
print(s.lower())        # "hello world"
print(s.upper())        # "HELLO WORLD"
print(s.swapcase())     # "HELLO WORLD" -> "hello world" (swap cases)
print(s.title())        # "Hello World"
```
